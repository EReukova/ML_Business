```python
#среднее значение, представление по неделям 
brent_data_W_m = brent_data.resample('W').mean()
brent_data_W_m.plot(figsize=(12,6))
plt.ylabel('USA dollar')
plt.xlabel('t')
```

```python
#среднее значение, представление по месяцам 
brent_data_M_m = brent_data.resample('M').mean()
brent_data_M_m.plot(figsize=(12,6))
plt.ylabel('USA dollar')
plt.xlabel('t')
```

```python
#среднее значение, представление по годам
brent_data_Y_m = brent_data.resample('Y').mean()
brent_data_Y_m.plot(figsize=(12,6))
plt.ylabel('USA dollar')
plt.xlabel('t')
```

```python
#среднее отклонение, представление по неделям 
brent_data_W_s = brent_data.resample('W').std()
brent_data_W_s.plot(figsize=(12,6))
plt.ylabel('USA dollar')
plt.xlabel('t')
```

```python
#среднее отклонение, представление по месяцам
brent_data_M_s = brent_data.resample('M').std()
brent_data_M_s.plot(figsize=(12,6))
plt.ylabel('USA dollar')
plt.xlabel('t')
```

```python
#среднее отклонение, представление по годам
brent_data_Y_s = brent_data.resample('Y').std()
brent_data_Y_s.plot(figsize=(12,6))
plt.ylabel('USA dollar')
plt.xlabel('t')
```
